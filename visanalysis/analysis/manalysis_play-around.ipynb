```python
import h5py
import numpy as np
import functools

import imaging_data
```

```python
file_path = '/Users/manzezhang/Documents/Stanford/Lab_Code/bruker_sr_data/25-11-04.hdf5'
with h5py.File(file_path, "r") as experiment_file:
    def print_structure(name, obj):
        print(f"{name}: {type(obj)}")

    experiment_file.visititems(print_structure)
```

```python
def find_series(name, obj, sn):
    target_group_name = 'series_{}'.format(str(sn).zfill(3)) # returns 'series_003' for sn=3 zfills means to pad with zeros to length 3
    if target_group_name in name:
        return obj
```

```python
with h5py.File(file_path, "r") as experiment_file:
    find_partial = functools.partial(find_series, sn=1)
    epoch_run_group = experiment_file.visititems(find_partial)
    print(epoch_run_group)            # <class 'h5py._hl.group.Group'>
```
